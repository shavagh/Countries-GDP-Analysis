```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/countries_gdp.csv')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.options.display.float_format = '{:.2f}'.format
df.head()
```

```python
####### CLEAN DATA #######
```

```python
df.drop([0, 1, 2], inplace=True)
cols = df.columns = df.iloc[0]
df.columns = df.iloc[0]
df.drop([3], inplace=True)
df.rename_axis('', axis=1, inplace=True)
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
df.shape
```

```python
intyears = [1960, 1961, 1962, 1963, 1964, 1965, 1966, 1967, 1968, 1969, 1970, 1971, 1972, 1973, 1974,
            1975, 1976, 1977, 1978, 1979, 1980, 1981, 1982, 1983, 1984, 1985, 1986, 1987, 1988, 1989,
            1990, 1991, 1992, 1993, 1994, 1995, 1996, 1997, 1998, 1999, 2000, 2001, 2002, 2003, 2004,
            2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]

stryears = ['1960', '1961', '1962', '1963', '1964', '1965', '1966', '1967', '1968', '1969', '1970', '1971', '1972', '1973', '1974',
            '1975', '1976', '1977', '1978', '1979', '1980', '1981', '1982', '1983', '1984', '1985', '1986', '1987', '1988', '1989',
            '1990', '1991', '1992', '1993', '1994', '1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004',
            '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019']
```

```python
df.drop(columns=['Indicator Name','Indicator Code'], inplace=True)
condition = np.array(intyears)
col_years = np.extract(condition, intyears)
col_country = np.array(['Country Name', 'Country Code'])
cols = np.append(col_country, col_years)
df.columns = cols   
df.head(10)
```

```python
df = pd.melt(df, id_vars=['Country Name', 'Country Code'], value_vars=stryears,
             value_name='GDP', var_name='Year').sort_values(['Country Name', 'Year'])
cols = df.columns
df.reset_index(drop=True, inplace=True)
df.head(20)
```

```python
regions = ['EUU', 'HIC', 'OED', 'PST', 'MEA', 'IBT', 'LMY', 'MIC', 'IBD', 'EAS',
           'UMS', 'NAC', 'LTE', 'ECS', 'EAP', 'TEA', 'EMU', 'EAR', 'LMC', 'LCN',
           'TLA', 'LAC', 'TEC', 'MEA', 'SAS', 'TSA', 'ECA', 'ARB', 'IDA', 'TSS',
           'SSF', 'SSA', 'CEB', 'PRE', 'IDX', 'LDC', 'IDB', 'HPC', 'SST', 'LIC',
           'OSS', 'FCS', 'TMN', 'MNA', 'INX', 'UMC', 'WLD']
regions_df = df[df['Country Code'].isin(regions)]
regions_df = regions_df.sort_values('Country Name', ascending=True).reset_index(drop=True)
region_indexes = df.index[df['Country Code'].isin(regions).tolist()]
regions_df.rename(columns={'Country Name': 'Region Name','Country Code': 'Region Code'}, inplace=True)

countries_df = df
countries_df.drop(region_indexes, inplace=True)
countries_df.reset_index(drop=True, inplace=True)
```

```python
# Top 20 GDP Countries
```

```python
top_country_2019 = countries_df[countries_df['Year']=='2019']
top_country_2019 = top_country_2019.sort_values('GDP', ascending=False).head(20)
top_country_2019.style.background_gradient(cmap='Blues')
```

```python
plt.figure(figsize=(8,10))
sns.barplot(x='GDP',y='Country Name',data=top_country_2019,palette='hsv_r')
plt.title('TOP 20 COUNTRIES GDP')
plt.show()
```

```python
top_countries_gdp = top_country_2019['Country Name'].tolist()
top_countries_gdp = countries_df[countries_df['Country Name'].isin(top_countries_gdp)]
top_countries_gdp
```

```python
plt.figure(figsize=(15,8))
sns.lineplot(x='Year',y='GDP',hue='Country Name',data=top_countries_gdp,palette='hsv')
plt.grid()
plt.title('TOP 20 COUNTRIES GDP OVERTIME')
plt.show()
```

```python
# Income groups overtime
```

```python
hic_region = regions_df[regions_df['Region Code']=='HIC'] 
lic_region = regions_df[regions_df['Region Code']=='LIC'] 
lmy_region = regions_df[regions_df['Region Code']=='LMY'] 
lmc_region = regions_df[regions_df['Region Code']=='LMC'] 
income_groups = [hic_region, lic_region, lmy_region, lmc_region] 
income_df = pd.concat(income_groups) 

plt.figure(figsize=(15,8)) 
sns.lineplot(x='Year',y='GDP',data=income_df,hue='Region Name',style='Region Name',markers=['*','o','<','>']) 
plt.grid() 
plt.title('INCOME GROUP GDP VARIATION OVERTIME') 
plt.show()
```
